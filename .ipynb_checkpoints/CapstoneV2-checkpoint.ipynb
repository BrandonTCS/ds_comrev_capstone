## Pre-processing part 2
### Cleaning the text of unecessary words or symbols, lower case all the words

```python
# helper function to clean the data for 

def clean_text(text):
    # remove the text that is not part of the actual comments
    pattern1 = re.compile(r'\nBe\sthe\sfirst.+')
    text = pattern1.sub(r'',text)
    pattern2 = re.compile(r'\d\speople\sfound\sthis.+')
    text = pattern2.sub(r'',text)
    pattern3 = re.compile(r'\d\sperson\sfound.+')
    text = pattern3.sub(r'',text)
    pattern4 = re.compile(r'\nHelpful\nShare\nRepor')
    text = pattern4.sub(r'',text)
    
    # remove "/n"
    pattern5 = re.compile(r'\n')
    text = pattern5.sub(r'',text)
    # remove irrelevant character such as numbers and punctuation
    pattern6 = re.compile(r'[^a-zA-Z0-9]')
    text = pattern6.sub(r' ',text)
    # remove extra whitespaces
    pattern7= re.compile(r'^\s*|\s\s*')
    text = pattern7.sub(r' ', text).strip()
    # text to all lower case
    text = text.lower()
    
    return text
```

```python
# running the helper function to clean both google and tencent dataset

tencent_df['pros'] = tencent_df['pros'].apply(lambda x: clean_text(str(x)))
google_df['pros'] = google_df['pros'].apply(lambda x: clean_text(str(x)))

tencent_df['cons'] = tencent_df['cons'].apply(lambda x: clean_text(str(x)))
google_df['cons'] = google_df['cons'].apply(lambda x: clean_text(str(x)))
```

### Tokenizing, removing stopwords, lemmatizing

```python
pd.set_option('display.max_colwidth', 999)
```

```python
# load spacy small english models
nlp = spacy.load('en_core_web_sm')
```

```python
# function to remove stopwords
def remove_stopwords(text):
    text = [token for token in text if not token.is_stop]
    return text

# function to lemmatize
def lemmatize(text):
    text = [token.lemma_ for token in text]
    return text

# combine function to perform both removal stopwords and lemmatizing
def combine(text):
    text = remove_stopwords(text)
    text = lemmatize(text)
    return text
```

```python
'''working on the pros column of the dataframe of both companies'''

# applying the spacy model to tokenize the sentence
google_df['pros'] = google_df['pros'].apply(lambda x: nlp(x))
tencent_df['pros'] = tencent_df['pros'].apply(lambda x: nlp(x))

# using the combine to do removal stopwords and lemmatizing
google_df['pros'] = google_df['pros'].apply(lambda x: combine(x))
tencent_df['pros'] = tencent_df['pros'].apply(lambda x: combine(x))

# list to string
google_df['pros'] = google_df['pros'].apply(lambda x: ' '.join(x))
tencent_df['pros'] = tencent_df['pros'].apply(lambda x: ' '.join(x))

'''working on the cons column of the dataframe of both companies'''

# applying the spacy model to tokenize the sentence
google_df['cons'] = google_df['cons'].apply(lambda x: nlp(x))
tencent_df['cons'] = tencent_df['cons'].apply(lambda x: nlp(x))

# using the combine to do removal stopwords and lemmatizing
google_df['cons'] = google_df['cons'].apply(lambda x: combine(x))
tencent_df['cons'] = tencent_df['cons'].apply(lambda x: combine(x))

# list to string
google_df['cons'] = google_df['cons'].apply(lambda x: ' '.join(x))
tencent_df['cons'] = tencent_df['cons'].apply(lambda x: ' '.join(x))
```

```python
'''cleaning the data after combining them for a final overview of the common words'''
def clean_2(text):
    
    # remove the spaces more than 1
    pattern = re.compile(r'\s{2,}')
    text = pattern.sub(r'',text)
    # remove the start of the string
    pattern1 = re.compile(r'^.')
    text = pattern1.sub(r'',text)
    # remove the /n with numbers
    pattern2 = re.compile(r'\n\d+')
    text = pattern2.sub(r' ',text)
    # remove the /n with numbers
    pattern3 = re.compile(r'\nName:.+\w+')
    text = pattern3.sub(r'',text)
    return text
```

```python
'''using the clean functiont to process all the text above'''

'''Tencent pros & cons'''

pros_tencent = clean_2(pros_tencent)
cons_tencent = clean_2(cons_tencent)

'''Google pros & cons'''
pros_google = clean_2(pros_google)
cons_google = clean_2(cons_google)
```

## Modelling

### Feature Engineering

```python
# combine pros and cons
tencent_df['test'] = tencent_df['pros'] + ' ' + tencent_df['cons']
```

```python
tencent_df
```

```python
X_tencent = tencent_df['test']
y_tencent = tencent_df['rating_overall']

X_train, X_test, y_train, y_test = train_test_split(X_tencent, y_tencent, random_state=42, stratify=y_tencent)
```

```python
# Count Vec as features
countv = CountVectorizer(ngram_range=(1,3))
countv.fit_transform(X_tencent)

X_train_count = countv.transform(X_train)
X_test_count = countv.transform(X_test)
```

```python
# tfidf vec as features
tfidfv = TfidfVectorizer(ngram_range=(1,3))
tfidfv.fit(X_tencent)

X_train_tfidf = tfidfv.transform(X_train)
X_test_tfidf = tfidfv.transform(X_test)
```

### Naive Bayes Model

```python
# helper function to train model

def train_model(classifier, X_train, y_train, X_test):
    # fit the training dataset on the classifier
    classifier.fit(X_train, y_train)

    # predict the labels on validation dataset
    predictions = classifier.predict(X_test)

    return accuracy_score(predictions, y_test), plot_confusion_matrix(classifier,X_test,y_test,cmap='Blues',values_format='d')
```

```python
# Naive Bayes with count vec
train_model(MultinomialNB(), X_train_count, y_train, X_test_count)
```

```python
# Naive Bayes with tfidf
train_model(MultinomialNB(), X_train_tfidf, y_train, X_test_tfidf)
```

### Support Vector Machine

```python
# SVC with count vec
train_model(SVC(), X_train_count, y_train, X_test_count)
```

```python
# SVC with count vec
train_model(SVC(), X_train_tfidf, y_train, X_test_tfidf)
```

### Random Forest

```python
# Random Forest with count vec
train_model(RandomForestClassifier(), X_train_count, y_train, X_test_count)
```

```python
# Random Forest with count vec
train_model(RandomForestClassifier(), X_train_tfidf, y_train, X_test_tfidf)
```
